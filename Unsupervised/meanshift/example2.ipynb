```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift,estimate_bandwidth
```

```python
data, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)
```

```python
bandwidth = estimate_bandwidth(data, quantile=0.2, n_samples=500)
# quantile - Should be between [0, 1] 0.5 means that the median of all pairwise distances is used.
# n_samples - The number of samples to use. If not given, all samples are used.
```

```python
model = MeanShift(bandwidth=bandwidth)
model.fit(data)
```

```python
cluster_centers = model.cluster_centers_
```

```python
labels = model.labels_
```

```python
labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("number of estimated clusters : %d" % n_clusters_)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='rainbow')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], color='black', marker='X', s=100)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Mean-Shift Clustering')
plt.show()
```
